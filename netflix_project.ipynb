```python
'''                                                   Netfilx Data Analysis                           '''

# Step-1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step-2: Load Dataset
df = pd.read_csv('netflix_titles.csv')
df.columns = df.columns.str.strip()

# Step-3: Clean Data
df = df.dropna(subset=['type', 'release_year', 'rating', 'country', 'duration'])

```

```python
# Step-4: Data Analysis

# 1. Movies vs TV Shows : distinguish which type of content
type_counts = df['type'].value_counts()
plt.figure(figsize=(5, 3))
plt.bar(type_counts.index, type_counts.values, color=['aqua', 'violet'], width=0.2)
plt.title('Distribution of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 2. Rating Counts
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Content Rating Distribution')
plt.tight_layout()
plt.show()

```

```python
# 3. Movie Duration
movies = df[df['type'] == 'Movie'].copy()
movies['duration_int'] = movies['duration'].str.replace('min', '').astype(int)
plt.figure(figsize=(10, 4))
plt.hist(movies['duration_int'], bins=20, color='navy', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

```python
# 4. Release Year Trend
release_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10, 4))
plt.plot(release_counts.index, release_counts.values, marker='o', color='green')
plt.title('Titles Released Per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 5. Top 10 Countries
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.barh(top_countries.index, top_countries.values, color='lime')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

```

```python
# 6. Year-wise Content by Type
content_by_year = df.groupby(['release_year', 'type']).size().unstack().fillna(0)
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
ax[0].plot(content_by_year.index, content_by_year['Movie'], color='orange')
ax[0].set_title('Movies per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')

ax[1].plot(content_by_year.index, content_by_year['TV Show'], color='grey')
ax[1].set_title('TV Shows per Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Count')

plt.suptitle('Movies vs TV Shows Over the Years')
plt.tight_layout()
plt.show()

```

```python
# 7. NEW: Top 10 Directors
top_directors = df['director'].dropna().value_counts().head(10)
plt.figure(figsize=(10, 5))
plt.barh(top_directors.index[::-1], top_directors.values[::-1], color='lime')
plt.title('Top 10 Directors by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.tight_layout()
plt.show()
```

```python
# 8. NEW: Top 15 Countries Bar Chart
top_15_countries = df['country'].value_counts().head(15)
plt.figure(figsize=(12, 6))
plt.bar(top_15_countries.index, top_15_countries.values, color='steelblue')
plt.xticks(rotation=45)
plt.title('Content Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```
